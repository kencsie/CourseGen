```python
from langchain_community.document_loaders.parsers import RapidOCRBlobParser
from langchain_community.document_loaders import PyMuPDFLoader


loader = PyMuPDFLoader(
    "../data/Prompt Engineering_v7_Google.pdf",
    mode="page",
)
docs = loader.load()
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",                 # 或其他 sentence-transformers 模型
    model_kwargs={"device": "cpu"},          # 沒 GPU 就用 "cpu"
    encode_kwargs={"normalize_embeddings": True},
)
```

```python
from langchain_chroma import Chroma
vs = Chroma(
    collection_name="pdf_kb",
    embedding_function=embeddings,
    persist_directory="./chroma_langchain_db",
)
```

```python
vs.add_documents(docs)
retriever = vs.as_retriever(search_type="mmr", search_kwargs={"k": 6, "fetch_k": 30})
```

```python
retriever.batch(["few-shot"])
```
